```python
from transformers import MarianMTModel, MarianTokenizer
from nltk.translate.bleu_score import sentence_bleu
```

```python
model_name="Helsinki-NLP/opus-mt-en-fr"
model=MarianMTModel.from_pretrained(model_name)
tokenizer=MarianTokenizer.from_pretrained(model_name)
```

```python
!pip install transformers
```

```python
src_texts=['hello, how are you?']
translated=tokenizer(src_texts, return_tensors='pt', padding=True)
translated_ids=model.generate(**translated)
tgt_texts=[tokenizer.decode(t, skip_special_tokens=True) for t in translated_ids]
print(tgt_texts)
```

```python
translated_ids=model.generate(**translated, num_beams=5, early_stopping=True)
```

```python
print(translated_ids)
```

```python
translated_text=tokenizer.decode(translated_ids[0], skip_special_tokens=True)
```

```python
print('translated_text:', translated_text)
```

```python
reference = [['Bonjour', 'comment', 'ça', 'va']]  # Should match the target sentence
```

```python
hypothesis=translated_text.split()
bleu_score=sentence_bleu(reference,hypothesis, weights=(0.25, 0.25, 0.25, 0.25))
print(bleu_score)
```
